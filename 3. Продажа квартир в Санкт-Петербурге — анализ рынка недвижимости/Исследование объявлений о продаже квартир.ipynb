## Расчёты и добавление результатов в таблицу

Посчитаем и добавим в таблицу:  
- цену квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры; варианты — первый, последний, другой;
- соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
# Цена за метр2

data['price_m2'] = data['last_price'] / data['total_area_m2']
data['price_m2'] = data['price_m2'].round(decimals=2)
data.head()
```

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format = '%Y-%m-%d')
```

```python
# Работаем с датой

data['announcement_day'] = data['first_day_exposition'].dt.day_name()
data['announcement_month'] = data['first_day_exposition'].dt.month_name()
data['announcement_year'] = data['first_day_exposition'].dt.year
data.head()
```

```python
# Добавим этажи

def floor_house(i):
    if i['floor'] == 1:
        return 'первый'
    elif i['floor'] == i['total_floors_in_house']:
        return 'последний'
    return 'другой'
```

```python
data['floor_house'] = data.apply(floor_house, axis=1)
data.head()
```

```python
# Добавим соотношение жилой к общей площади и площадь кухни к общей площади

data['total_living'] = data['living_area_m2'] / data['total_area_m2']
data['kithen_to_total'] = data['kitchen_area_m2'] / data['total_area_m2']
data.head()
```
